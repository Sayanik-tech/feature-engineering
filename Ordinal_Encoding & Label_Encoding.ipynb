```python
x = df.iloc[:,:4]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
X_train.shape
```

```python
X_train
```

```python
## Ordinal Encoding
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Poor', 'Average','Good'],['School', 'UG', 'PG']])
```

```python
df_ordinal_train = pd.DataFrame(oe.fit_transform(X_train[['review','education']]),columns=['review','education'])
df_ordinal_test = pd.DataFrame(oe.transform(X_test[['review','education']]),columns=['review','education'])
df_ordinal_train.reset_index(drop=True,inplace=True)
df_ordinal_test.reset_index(drop=True,inplace=True)
```

```python
df_ordinal_train
```

```python
df_ordinal_test
```

```python
## Nominal Ctegorical Variable encoding
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first',sparse=False,dtype=np.int32)
```

```python
df_nominal_train = pd.DataFrame(ohe.fit_transform(X_train[['gender']]),columns=['gender'])
df_nominal_test = pd.DataFrame(ohe.transform(X_test[['gender']]),columns=['gender'])
df_nominal_train.reset_index(drop=True,inplace=True)
df_nominal_test.reset_index(drop=True,inplace=True)
```

```python
X_train['age'].reset_index(drop=True,inplace=True)
X_test['age'].reset_index(drop=True,inplace=True)
```

```python
X_train_final= pd.concat([X_train['age'],df_nominal_train,df_ordinal_train],axis = 1)
```

```python
X_test_final= pd.concat([X_test['age'],df_nominal_test,df_ordinal_test],axis=1)
```
