```python
import pandas as pd
import numpy as np
```

```python
date = pd.read_csv('E:/Dataset/orders.csv')
time = pd.read_csv('E:/Dataset/messages.csv')
```

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
time['date'] = pd.to_datetime(time['date'])
```

```python
## Extract year and Month,month name,day

date['date_year'] = date['date'].dt.year
date['date_month'] = date['date'].dt.month
date['month_name'] = date['date'].dt.month_name()
date['day'] = date['date'].dt.day
```

```python
date.sample(5)
```

```python
## Which day of the week
date['day_of_week'] = date['date'].dt.dayofweek
```

```python
date.head()
```

```python
date['day_dow'] = date['date'].dt.day_name()
```

```python
date.sample(5)
```

```python
date.drop(columns=['product_id','city_id','orders'],axis=1,inplace=True)
```

```python
date.sample(5)
```

```python
## Is the weekend or not

date['Weekend'] = np.where(date['day_dow'].isin(['Saturday','Sunday']),1,0)
```

```python
date.head()
```

```python
## extract week of the year
date['week_number'] = date['date'].dt.week
```

```python
date.sample(5)
```

```python
# Extracting Quarter
date['quarter'] = date['date'].dt.quarter
```

```python
date.sample(5)
```

```python
# Extracting Semester

date['Semester'] = np.where(date['quarter'].isin([1,2]),1,2)
```

```python
date.sample(5)
```
