```python
df_country = pd.read_excel('zomato_dataset/Country-Code.xlsx')
```

```python
df_country.head()
```

```python
df_final = pd.merge(df,df_country,how='left',on='Country Code')
```

```python
df_final.head(2)
```

```python
# to check the datatypes
df_final.dtypes
```

```python
df_final.Country.value_counts()
```

```python
##Most of the transactions of zomato is generated from India
```

```python
country_name = df_final.Country.value_counts().index
```

```python
country_values = df_final.Country.value_counts().values
```

```python
# pie chart plotting --- for top 3 countries as per Transactions

plt.pie(country_values[:3],labels=country_name[:3],autopct='%1.2f%%')
```

observations:Zomatos most of the transactions are from India ,,then USA ,, and then UK

```python
df_final.columns
```

```python
df_final.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
df_final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
ratings = df_final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

Observations:
1.When rating is between 4.5 to 4.9 ---->Excellent
1.When rating is between 4.0 to 4.4 ---->Very Good
1.When rating is between 3.5 to 3.9 ---->Good
1.When rating is between 2.5 to 3.4 ---->Average
1.When rating is between 1.8 to 2.4 ---->Poor
1.When rating is between 0.0 ---->Not Rated

```python
# barlot
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings)
```

```python
sns.barplot(x = "Aggregate rating", y ="Rating count",hue='Rating color',data = ratings,palette=['Grey','red','orange','yellow','green','green'])
```

Observations:
1.Not rated count(Grey in colour) is very high
2.Maximum no of count is between 2.5 to 3.4

```python
# Find the countries that gives 0 rating

df_final.loc[df_final['Aggregate rating']== 0.0,['Country','Aggregate rating']].value_counts().reset_index().rename(columns={0:'Counts'})
```

Observations:Most of the 0 ratings from Indian Customers
